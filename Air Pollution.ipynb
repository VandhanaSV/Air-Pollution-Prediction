# Air Pollution

Air pollution refers to the presence of harmful or excessive quantities of substances in the air that can have adverse effects on human health, the environment, and the quality of life. These substances, known as air pollutants, can be natural or human-made and can vary in their composition and sources.

# Description


country_name: This column contains the name of the country where the air quality data was recorded.
city_name: This column contains the name of the city where the air quality data was recorded.
aqi_value: AQI stands for Air Quality Index. This column contains the numeric value representing the overall air quality index at the given location. It is a standardized way of quantifying and communicating air pollution levels to the public.
aqi_category: This column provides a categorical classification of the air quality based on the AQI value. It typically ranges from "Good" to "Hazardous" and may include categories such as "Moderate", "Unhealthy", "Very Unhealthy", and "Hazardous", among others.
co_aqi_value: This column represents the AQI value specifically for carbon monoxide (CO) levels.
co_aqi_category: Similar to the "aqi_category" column, this column provides a categorical classification of the carbon monoxide AQI value.ozone_aqi_value: This column represents the AQI value specifically for ozone (O3) levels.
ozone_aqi_category: Similar to the "aqi_category" column, this column provides a categorical classification of the ozone AQI value.
no2_aqi_value: This column represents the AQI value specifically for nitrogen dioxide (NO2) levels.
no2_aqi_category: Similar to the "aqi_category" column, this column provides a categorical classification of the nitrogen dioxide AQI value.
pm2.5_aqi_value: This column represents the AQI value specifically for particulate matter (PM2.5) levels. PM2.5 refers to fine particles with a diameter of 2.5 micrometers or smaller, which can penetrate deep into the lungs.
pm2.5_aqi_category: Similar to the "aqi_category" column, this column provides a categorical classification of the PM2.5 AQI value.
target value-aqi_value

```python
b['country_name']=d.fit_transform(b['country_name'])
b
```

# Prediction - Regression

```python
x=b.iloc[:,[0,1,3,4,5,6,7,8,9,10,11]].values #independent values
x
```

```python
y=b.iloc[:,2].values  #dependent value
y
```

Train_Test_Split

```python
from sklearn.model_selection import train_test_split 
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=60,random_state=10)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

# Linear

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor=LinearRegression()
regressor
```

```python
regressor.fit(x_train,y_train)
```

```python
xpred1=regressor.predict(x_train)
xpred1
```

```python
ypred1=regressor.predict(x_test)
ypred1
```

```python
trainingscore1=regressor.score(x_train,y_train)
trainingscore1
```

```python
testingscore1=regressor.score(x_test,y_test)
testingscore1
```

```python
from sklearn import metrics
```

```python
print("mae"),metrics.mean_absolute_error(y_test,ypred1) #mean absolute error
```

```python
print("mse"),metrics.mean_squared_error(y_test,ypred1)  #mean squared error
```

```python
print("rmae"),np.sqrt(metrics.mean_absolute_error(y_test,ypred1)) #mean absolute error
```

```python
print("rmse"),np.sqrt(metrics.mean_squared_error(y_test,ypred1)) #mean squared error
```

```python
from sklearn.metrics import r2_score
```

```python
r1=r2_score(y_test,ypred1)
r1
```
